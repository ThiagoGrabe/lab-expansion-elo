de  qualquer  sistema  de  saúde  em  qualquer  lugar  do  mundo,  sendo  somente  nos  EUA  um

### O Problema

Diante desse cenário, o  desafio proposto tem como objetivo auxiliar a  área de  negócios de 
uma  rede norte americana de medicina diagnóstica a ter insumos iniciais para suportar seu 
plano de expansão da rede de laboratórios. Sabe-se que não é possível ou é bem difícil gerar 
demanda e fazer ofertas neste tipo de mercado, entretanto é muito interessante e rentável 
fidelizar o cliente para que ele sempre visite a rede quando surgir a necessidade.

```python
DATA_FOLDER = '../data/'
print('\nData Folder:')
filenames = os.listdir(DATA_FOLDER)
for filename in filenames:
    if filename.endswith( 'csv' ):
        print(filename)
print('\nDemographic_data_ZCTAs Folder:') 
filenames = os.listdir(os.path.join(DATA_FOLDER,'Demographic_data_ZCTAs'))
for filename in filenames:
    if filename.endswith( 'csv' ):
        print(filename)
print('\nEconomic_data_ZCTAs Folder:') 
filenames = os.listdir(os.path.join(DATA_FOLDER,'Economic_data_ZCTAs'))
for filename in filenames:
    if filename.endswith( 'csv' ):
        print(filename)
print('\nHealth_insurance_data_ZCTAs Folder:') 
filenames = os.listdir(os.path.join(DATA_FOLDER,'Health_insurance_data_ZCTAs'))
for filename in filenames:
    if filename.endswith( 'csv' ):
        print(filename)
```

```python
# Geo Code dataset
df_geocode = pd.read_csv(os.path.join(DATA_FOLDER, 'df_geocode.csv'), delimiter=';')
df_geocode[['lat', 'long']] = df_geocode['location'].str.split(',', 1, expand=True)
df_geocode.drop(df_geocode[df_geocode['Address'].str.contains("Unavailable")==True].index, inplace=True)
df_geocode['lat'] = pd.to_numeric(df_geocode['lat'])
df_geocode['long'] = pd.to_numeric(df_geocode['long'])
df_geocode.head()
```

```python
uszips = pd.read_csv(os.path.join(DATA_FOLDER, 'uszips.csv'))
uszips.head()
```

```python
usa_map = gpd.read_file(os.path.join(DATA_FOLDER, 'cb_2020_us_county_5m.shp'), bbox=None, mask=None, rows=None)
fig,ax = plt.subplots(figsize = (15,15))
usa_map.plot(ax = ax)
geometry = [Point(xy) for xy in zip(df_geocode.long,df_geocode.lat)]
geo_df = gpd.GeoDataFrame(geometry = geometry)
g = geo_df.plot(ax = ax, markersize = 20, color = 'red',marker = '*')
ax.set_title('Localização dos Laboratórios', fontsize=25)
ax.text(x=-120, y=20, s='Laboratórios com posições diversificadas\ncom destaque para as regiões próximas ao litoral.', fontsize=15)
ax.axis('Off')
ax.set_xlim(-125,-65)
ax.set_ylim(20,50)
plt.show()
```

```python
# Test dataset
df_testdata = pd.read_csv(os.path.join(DATA_FOLDER, 'test_data.csv'), delimiter=';')
df_testdata['Testing Cost'] = df_testdata['Testing Cost'].str.replace(',','.')
df_testdata['Testing Cost'] = pd.to_numeric(df_testdata['Testing Cost'])
# df_trans.duplicated(subset=['CodItem']).any()
df_testdata.head()
```

```python
# Transactional dataset
df_trans = pd.read_csv(os.path.join(DATA_FOLDER, 'transactional_data.csv'), delimiter=';')
df_trans['Testing Cost'] = df_trans['Testing Cost'].str.replace(',','.')
df_trans['Testing Cost'] = pd.to_numeric(df_trans['Testing Cost'])
df_trans['Date of service'] = pd.to_datetime(df_trans['Date of service'])
df_trans['Date of birth'] = pd.to_datetime(df_trans['Date of birth'], dayfirst=True)
df_trans = df_trans.drop_duplicates(keep='last')
df_trans.head()
```

```python
df = pd.merge(left=df_trans, right=df_testdata, left_on='CodItem', right_on='CodItem',how='inner').sort_values(by=['Patient Id', 'Service Id', 'Date of service']).reset_index(drop=True)
df = pd.merge(left=df, right=df_geocode, left_on='Lab Id', right_on='Lab Id',how='inner').sort_values(by=['Patient Id', 'Service Id', 'Date of service']).reset_index(drop=True)
df = pd.merge(left=df, right=uszips[['zip','city','state_name', 'county_name']], left_on='zipcode', right_on='zip',how='inner').sort_values(by=['Patient Id', 'Service Id', 'Date of service']).reset_index(drop=True)

df['lat'] = pd.to_numeric(df['lat'])
df['long'] = pd.to_numeric(df['long'])
print(df.shape)
df.rename(columns={'Testing Cost_x' : 'Patient Cost', 'Testing Cost_y': 'Lab Cost'}, inplace=True)
df.drop(df.loc[df['Lab Id']=='-2'].index, inplace=True)
df.drop(df.loc[df['Patient Id']=='-2'].index, inplace=True)
df['Estimated Profit'] = df['Patient Cost'] - df['Lab Cost']
print(df.shape)
df.head()
```
